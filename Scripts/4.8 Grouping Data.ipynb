## 04. New column for "max_order" created to sort users by max amount ordered

```python
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
df.head(20)
```

## 04. Loyalty flag created to classify how loyal customer are depending on max_order amount

```python
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df.loc[(df['max_order'] <= 40 ) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

## 06 Create a spending flag for different users based on their average amount spent

```python
# Create a variable for average spent calculation for each user and store it in new a column 
```

```python
df['avg_spent'] = df.groupby(['user_id'])['prices'].transform(np.mean)
df.head(10)
```

```python
# Create if/then conditions based on prices from avg_spent
```

```python
df['avg_spent'].isnull().sum()
```

```python
(df['avg_spent'] == 0).sum()
```

```python
df.loc[df['avg_spent'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
df.loc[df['avg_spent'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
df['spending_flag'].value_counts()
```

```python
df[['user_id', 'avg_spent', 'spending_flag']].head(10)
```

```python
# A majority of our customers are low spenders
# Instacart should focus target of campaign on these spenders
```

## 07 Create a frequency flag for different users depending on how often they ordered

```python
# Create a variable for the median amount of days since prior order and store it in new column
```

```python
df['med_days_since_order'] = df.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
# Create if/then calculations based on median days 
```

```python
df.loc[df['med_days_since_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df.loc[(df['med_days_since_order'] > 10) & (df['med_days_since_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df.loc[df['med_days_since_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df['frequency_flag'].value_counts()
```

```python
df[['user_id', 'med_days_since_order', 'frequency_flag']].head(10)
```
